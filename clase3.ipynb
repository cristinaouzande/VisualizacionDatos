```python
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME' #todas las columnas
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = 'std'  #se le puede cambiar ese a sum, mean, median, std
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = ['std', 'sum', 'mean', 'median']  #para poner varias y se le puede cambiar ese a sum, mean, median, std
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['STNAME', 'CTYNAME'],
                          values = ['BIRTHS2010','BIRTHS2011'],
                          aggfunc = 'median', #se le puede cambiar ese a sum, mean, median, std
                          columns = 'SUMLEV',
                          margins = True 
                         )
my_pivot
```

```python
df = pd.read_csv('datasets/airline_data.csv', low_memory=False)
df.head()
```

```python
delay_pivot = pd.pivot_table(df,
                             index = ['Year','Quarter'],
                             values = 'DepDelay',
                             aggfunc = 'sum',
                             columns = 'DepartureDelayGroups',
                             margins = True
                            )
delay_pivot
```
