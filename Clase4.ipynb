## Pregunta de investigacion 1
### existe alguna correlacion entre las variables wingspan y cw distance? 


```python
sns.lmplot (x='Wingspan',
            y = 'CWDistance',
            data = df,
            fit_reg=False, #para quitar la linea de regresion
            hue = 'Gender'
           )
plt.title('Scatter plot of Wingspan vs CWDistance', fontsize = 17)
plt.xticks(fontsize=15,rotation=35)

plt.show()
```

## Conclusion: 
No se encontro ninguna correlacion entre las variables Wingspan y CWDistance
