## Descripción de las Variables de Interés 

Mis variables de interés en esta base de datos "Wine Quality" son: quality, pH, alcohol, y free sulfur dioxide.

- Quality: La calidad es el output variable tambien considerada una variable categorica ya que determina el nivel de calidad del vino tinto, se base en una escala del 0-10, 10 siendo la mejor calidad. 
- pH: El pH es una variable numérica, esta indica la cantidad de pH, la cual determina la acidez del vino. Un vino con un pH de 2.8 es extremadamente ácido, mientras que un vino con un pH de 4 no es ácido. 
- Alcohol: El alohol es una variable numérica ya que indica la cantidad de alcohol en el vino. 
- sulphates: Sulphates es una variable númerica, esta es una variante de oxido de azufre que se genera en el proceso de fermentación del vino.

## Pregunta 7: ¿Qué tan bueno es el análisis de regresión para la calidad en función de pH, alcohol, sulphates? ¿Que indica la R-cuadrada y los P-values de los parametros B0, B1, B2, B3? 

```python
model = sm.OLS.from_formula('quality ~  pH + alcohol + sulphates', data = df)
result = model.fit() 
result.summary()
```

### Observaciones: 

- Variable dependiente: Quality
- Variable independiente: pH, alcohol, sulphates
- R-squared: 0.283, el modelo no tan eficiente ya que el valor esta muy cercano a 0. Esto explica que solo el 28.3% de la calidad es basado en el pH, alcohol, y los sulphates.
