```python
df = pd.read_csv('antropometria-dataset-2.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df, 
                          index = 'desc_ent', 
                          values = 'peso',
                          aggfunc = 'mean'
                         )
my_pivot           
```

```python
my_pivot = pd.pivot_table(df, 
                          index = 'desc_ent', 
                          values = 'peso',
                          aggfunc = 'mean',
                          columns = 'sexo'
                         )
my_pivot    
```

Despues de realizar las tablas pivote para los datos de antropometria, puedo concluir que la informacion se ve mas ordenada de esta manera ya que podemos observar el peso promedio por estado en la primera tabla pivote, y el peso promedio por sexo en la segunda tabla pivote. Es un claro ejemplo de como podemos utilizar la programaciÃ³n para hacer un breve analisis y visualizaciÃ³n de nuestros datos.
